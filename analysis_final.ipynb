```python
import pandas as pd

df = pd.read_csv('final_aprox_27.csv', sep=';')

df[:6]
```

```python
df['dif_elem'] = abs(df['num_elem_1'] - df['num_elem_2'])

perc = 60
dif_elem = 4
cond_perc1 = (df['percent_1'] > perc)  #| True
cond_perc2 = (df['percent_2'] > perc)  #| True
cond_elem = (df['dif_elem'] <= dif_elem)  #| True

cond =  cond_perc1 & cond_perc2 & cond_elem
res = df.loc[ cond , :]
print(f'rinduri {len(res)}')

# res[:3]
res
```

```python
import subprocess
import os
path = f'aprox_blocks_perc > {perc}, dif_elem <= {dif_elem}, {len(res)} buc'
path = path.replace(' ','_')
os.mkdir(path)

# res
i = 1
for k, row in res.iterrows():
    way = path + '/' + str(i)
    os.mkdir(way)

    src1 = '/'.join(['blocks', row['block_1']])
    src1 = src1.replace("__", ",_")
#     print(row['block_1'])
    dst1 = '/'.join([way, row['block_1'].split('/')[0]+'--'+row['block_1'].split('/')[1]  ])
#     print(dst1)
    command1 = f"cp \'{src1}\' \'{dst1}\'"
    subprocess.call(command1, shell=True)
#     print(command1)
    
#     print(dst1)
#     print(way)  

    src2 = '/'.join(['blocks', row['block_2']])
    src2 = src2.replace("__", ",_")
    dst2 = '/'.join([way, row['block_2'].split('/')[0]+'--'+row['block_2'].split('/')[1]  ])
    command2 = f"cp \'{src2}\' \'{dst2}\'"
    subprocess.call(command2, shell=True)
    
#     print(dst1)
#     print(way)
#     print(command2)
    print('\n')
    i += 1
    
print('end code ...')
```
